## Handling Imbalanced Dataset

1. Up Sampling
2. Down Sampling

1. Up Sampling


```python
import pandas as pd
import numpy as np

np.random.seed(123)

n_sample = 1000
class_0_ratio = 0.9
class_0 = int(n_sample * class_0_ratio)
class_1 = n_sample - class_0
```

```python
class_0,class_1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
df_class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=class_0),
    'target': [0] * class_0
})

df_class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=class_1),
    'target': [1] * class_1
})
```

✔ feature_1

Es werden Zufallszahlen erzeugt
Die Werte stammen aus einer Normalverteilung (Gauß-Verteilung)
loc=0 → Mittelwert = 0
scale=1 → Standardabweichung = 1
size=n_class_0 → Anzahl der erzeugten Werte entspricht der Anzahl der Klasse-0-Samples

✔ feature_2

Gleiche Logik wie bei feature_1
Ebenfalls Normalverteilung mit Mittelwert 0 und Standardabweichung 1

✔ target

Dies ist die Zielspalte für maschinelles Lernen
[0] * n_class_0 erzeugt eine Liste voller Nullen
Beispiel: wenn n_class_0 = 100, dann wird daraus
[0, 0, 0, ..., 0] (100 mal)

Das bedeutet: alle Zeilen in diesem Dataset gehören zur Klasse 0.

```python
df = pd.concat([df_class_0, df_class_1]).reset_index(drop=True)
```

```python
df.tail(5)
```

```python
df['target'].value_counts()
```

```python
## upsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True, #Sample With replacement
         n_samples=len(df_majority),
         random_state=42
        )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

2. Dawn Sampliing

```python
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
## downsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample

df_majority_downsampled=resample(df_majority,replace=True, #Sample With replacement
         n_samples=len(df_minority),
         random_state=42
        )
```

```python
print(df_majority_downsampled)
```

```python
df_downsempled = pd.concat([df_minority,df_majority_downsampled])
print(df_downsempled.shape)
```

```python
# print(df_downsempled.value_counts)
df_downsempled['target'].value_counts()
```
